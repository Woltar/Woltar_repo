Загрузить нужные библиотеки и импортировать файл

```python
import pandas as pd
import numpy as np

df=pd.read_csv('ad.csv', delimiter=',')
print('Библиотеки и данные успешно загружены')
```

a. Выведите дескриптивную статистику для группировки по товару.

```python
df.groupby('ad_id').describe()
```

b. Сохраните в отдельную переменную и выведите значения поля action, которые соответствуют только товару A из всей представленной выборки

```python
x=np.array(df['action'][df['ad_id'].isin(['A'])])
x
```

c. Сохраните в отдельную переменную значения поля action, которые соответствуют только товару B из всей представленной выборки

```python
y=np.array(df['action'][df['ad_id'].isin(['B'])])
y
```

d. Подсчитайте среднее для переменных из заданий b и c.Сохраните получившиеся результаты в отдельные переменные. Выведите ответ по обеим получившимся переменным.

```python
xm = x.mean()
ym = y.mean()
print(f'Среднее по рекламе А = {xm}')
print(f'Среднее по рекламе B = {ym}')
```

e. Подсчитайте дисперсию для переменных из заданий b и c со степенью свободы равной 1 (ddof=1). Сохраните получившиеся результаты в отдельные переменные.Выведите ответ по обеим получившимся переменным.

```python
xd = x.var(ddof=1)
yd = y.var(ddof=1)

print(f'Дисперсия по рекламе А = {xd}')
print(f'Дисперсия по рекламе B = {yd}')
```

f. Добавьте новый атрибут new_ad_id в DataFrame на основании lambda функции по правилу “ЕСЛИ new_id = “B” ТОГДА 0 ИНАЧЕ 1”.

```python
df['new_ad_id'] = df['ad_id'].apply(lambda x: 1 if x == 'B' else 0)
df
```

g. Вызовите функцию корреляции между числовыми атрибутами DataFrame.

```python
df.corr(method='pearson')
```

h. Используя функцию probplot (пакета scipy.stats) постройте визуализацию для столбца new_ad_id по анализу его соответствия нормальному распределению.

```python
from scipy import stats
import matplotlib.pyplot as plt
import pylab

stats.probplot(df['new_ad_id'], dist='norm', plot=pylab)
pylab.show()
```
